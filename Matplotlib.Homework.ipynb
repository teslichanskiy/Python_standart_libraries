**7. Изобразить контурную карту функции $f(x, y) = \frac{\sin(x)}{\cos(y) + \pi/2}$ в области $[-3\pi, 3\pi] x [-2\pi, 2\pi]$; подписать оси $x, y$; отобразить цветовую шкалу [10]**

Полезные функции для решения задачи:

`np.linspace`, `plt.colormesh`, `np.colorbar`

```python
x = np.linspace(-3*np.pi, 3*np.pi, 500)
y = np.linspace(-2*np.pi, 2*np.pi, 500)
y = y.reshape(500, 1)
fxy = np.sin(x)/(np.cos(y) + np.pi/2)

plt.figure(figsize=(12,8))
plt.xlabel("x")
plt.ylabel("y")
f = plt.pcolormesh(x, y, fxy, cmap='twilight', shading='auto')
plt.colorbar(f)
plt.grid(ls='-.')
plt.title('$F(x, y) = \sin(x) / (\cos(y) + \pi/2)$');
plt.draw()
```

Полезные функции для решения задачи:

`np.linspace`, `plt.rcParams`, `plt.colormesh`, `np.colorbar`, `plt.xticks`, `plt.yticks`, `plt.title`

Для изменения засечек цветовой шкалы и подписей к ним:

```python
x = np.linspace(-5, 5, 512)
y = np.linspace(-1.5*np.pi, 1.5*np.pi, 512)
y = y.reshape(512, 1)
z = np.arctan(x)*np.cos(y)

plt.figure(figsize=(14,12))


plt.xticks(np.arange(-5, 6))

plt.yticks(np.arange(-np.pi, 1.5*np.pi, np.pi/2), [r'$-\pi$', r'$-\frac{\pi}{2}$', '0', r'$\frac{\pi}{2}$', r'$\pi$'])

a = plt.pcolormesh(x, y, z, cmap='jet', shading='auto')
plt.grid(color='k', linestyle=(0, (1,6)))

ax = plt.colorbar(a)
ax.set_ticks(np.arange(-np.pi/3, np.pi/2, np.pi/6))
ax.set_ticklabels([r'$-\frac{\pi}{3}$', r'$-\frac{\pi}{6}$', '0', r'$\frac{\pi}{6}$', r'$\frac{\pi}{3}$'])

plt.draw()
```

**9. Построить изображение функции, заданной неявно: $(x^2 + y^2)^2 - 2(x^2 - y^2) = 0$; подписать оси $x, y$  [10]**

Полезные функции для решения задачи:

`np.linspace`, `np.meshgrid`, `plt.contour`

```python
x = np.linspace(-2, 2, 200)
y = np.linspace(-1, 1, 200)

X, Y = np.meshgrid(x,y)

plt.contour(X, Y, ((X**2 + Y**2)**2 - 2*(X**2 - Y**2)), [0])
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid()
plt.draw()
```

**10. Построить [трехмерное](https://matplotlib.org/2.0.2/mpl_toolkits/mplot3d/tutorial.html) изображение [тора](https://ru.wikipedia.org/wiki/Тор_(поверхность)) при $R = 10$, $r = 4$; установить одинаковый масштаб по осям при помощи `ax.auto_scale_xyz` сохранить рисунок в файл `torus.png` с разрешением `300 dpi` [10]**

Параметрические уравнения тора:

$x(\phi, \psi) = (R + r\cos(\psi))\cos(\phi)$

$y(\phi, \psi) = (R + r\cos(\psi))\sin(\phi)$

$z(\phi, \psi) = r \sin(\psi)$

где $\phi \in [0, 2 \pi), \psi \in [-\pi, \pi)$

Полезные функции для решения задачи:

`np.linspace`, `np.meshgrid`, `ax.plot_surface`, `ax.auto_scale_xyz`, объект `Axes3D` из модуля `mpl_toolkits.mplot3d`

```python
from mpl_toolkits.mplot3d import Axes3D

R = 10
r = 4

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot((111), projection='3d')

phi = np.linspace(0, 2 * np.pi, 1000, endpoint=False)
psi = np.linspace(-np.pi, np.pi, 1000, endpoint=False)[:, None] #creates an axis with length 1

phi, psi = np.meshgrid(phi, psi)

x = (R + r * np.cos(psi)) * np.cos(phi)
y = (R + r * np.cos(psi)) * np.sin(phi)
z = r * np.sin(psi)

ax.plot_surface(x, y, z, cmap='Spectral')
bboxmin = np.min([x, y, z])
bboxmax = np.max([x, y, z])
ax.auto_scale_xyz([bboxmin, bboxmax], [bboxmin, bboxmax], [bboxmin, bboxmax])

fig.savefig('torus.png', dpi=300)
```
