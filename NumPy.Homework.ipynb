**6. Сгенерировать одномерные массивы `x, y`, содержащие координаты 360 точек, расположенных равномерно на единичной окружности. Объединить их в один двумерный массив размера (360, 2) [5]**

```python
import math
import matplotlib
import matplotlib.pyplot as plt
sin = np.vectorize(math.sin)
cos = np.vectorize(math.cos)
x = sin(np.linspace(0, 2 * math.pi, 360)).reshape(1, 360)
y = cos(np.linspace(0, 2 * math.pi, 360)).reshape(1, 360)
dots = np.transpose(np.concatenate((x, y)))
print(dots.shape)

linestyles = list(matplotlib.lines.Line2D.lineStyles)[:4]
plt.plot(x.reshape(360,1), y.reshape(360,1), ls=linestyles[0])

print(dots)
```

**7. Сгенерировать массив формы (100, 2), содержащий координаты точек, расположенных на плоскости случайно. Найти точки: ближайшую и самую дальнюю по отношению к началу координат [5]**

Полезные функции для решения задачи:

`np.linalg.norm`, `np.argmin`, `np.argmax`

```python
def norm(x): 
    return np.linalg.norm(x, ord=2)

x = np.array(np.random.randint(0, 100, 200)).reshape(100, 2)

leng = np.array(list(map(norm, x)))
z = np.argmin(leng)
print("Самая близкая точка ",x[z])
z = np.argmax(leng)
print("Самая дальная точка ",x[z])
```

**13. Сгенерировать массив `pts` размера (4, 2) из случайных чисел, равномерно распределенных в полуинтервале [0, 10). Считая, что массив хранит координаты точек на плоскости, провести через первую и вторую пару точек прямые и найти координаты точки их пересечения [5]**

Полезные функции для решения задачи:

`np.linalg.det`, `np.linalg.solve`

```python
pts = np.array(10 * np.random.random_sample((4,2)))
x1 = pts[0][1] - pts[1][1] 
y1 = pts[1][0] - pts[0][0] 
x2 = pts[2][1] - pts[3][1]
y2 = pts[3][0] - pts[2][0]
z1 = pts[1][0] * pts[0][1] - pts[0][0] * pts[1][1]
z2 = pts[3][0] * pts[2][1] - pts[2][0] * pts[3][1]

A = np.array([[x1, y1],[x2, y2]])
B = np.array([z1,z2])
if np.linalg.det(A) != 0:
    print(np.linalg.solve(A,B))
else:
    print("Имеется не одна точка пересечения прямых.(Прямые либо совпадают, либо параллельны)")
    
```

Найти коэфициенты $a$ и $b$ прямой, которая наилучшим образом аппроксимирует эти данные, используя метод наименьших квадратов (МНК). Проверить результаты при помощи функции `np.linalg.lstsq`. Изобразить результаты измерений и рассчитанную прямую на одном графике. [10]**

Согласно МНК, коэффициенты $a, b$ можно найти, решив систему уравнений:

$\begin{pmatrix}
\sum_{j=1}^{n}t_j^2 & \sum_{j=1}^nt_j\\
\sum_{j=1}^nt_j & \sum_{j=1}^n 1 
\end{pmatrix}\begin{pmatrix}a\\b\end{pmatrix}=\begin{pmatrix}\sum_{j=1}^{n}t_jx_j\\ \sum_{j=1}^{n}x_j\end{pmatrix}$
